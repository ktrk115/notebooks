```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
from ipywidgets import interact

img = cv2.imread('./data/Lenna.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def binarize(th):
    out = np.zeros(img.shape, dtype=np.uint8)
    out[np.where(img > th)] = 255
    plt.imshow(out, cmap=cm.Greys_r)
    
interact(binarize, th=(0,255))
```

```python
def edge(th_a, th_b):
    out = cv2.Canny(img, th_a, th_b)
    plt.imshow(out, cmap=cm.Greys_r)
interact(edge, th_a=(0,255), th_b=(0,255))
```
