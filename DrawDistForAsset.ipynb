```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")

names = ['Adelie', 'Chinstrap', 'Gentoo']
colors = ['#12711c', '#8c0800', '#001c7f']

plt.figure(figsize=(16,12))
ax = plt.subplot()
for name, color in zip(names, colors):
    df = penguins[penguins['species'] == name]
    cmap = sns.light_palette(color, as_cmap=True)
    sns.kdeplot(
        x=df['bill_length_mm'], y=df['bill_depth_mm'],
        cmap=cmap, fill=True, ax=ax, thresh=0.1, levels=6
    )
ax.set_axis_off()
```

```python
iris = sns.load_dataset("iris")
f, ax = plt.subplots(figsize=(16, 12))
cmap = sns.cubehelix_palette(as_cmap=True)
# cmap = sns.color_palette("flare", as_cmap=True)
sns.kdeplot(
    data=iris.query("species != 'versicolor'"),
    x="sepal_width", y="sepal_length",
    cmap=cmap, fill=True, ax=ax, thresh=0.05, levels=8
)
ax.set_axis_off()
```
