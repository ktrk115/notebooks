```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage.filters import sobel
from skimage.segmentation import felzenszwalb, slic, quickshift, watershed, mark_boundaries
from ipywidgets import interact

img = cv2.imread('./data/Lenna.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def sp_felzenszwalb(scale, sigma, min_size):
    seg = felzenszwalb(img, scale=scale, sigma=sigma, min_size=min_size)
    out = mark_boundaries(img, seg)
    plt.imshow(out)
interact(sp_felzenszwalb, scale=(0,200), sigma=(0,1.0), min_size=(0,100))
```

```python
def sp_slic(n_segments, compactness, sigma):
    seg = slic(img, n_segments=n_segments, compactness=compactness, sigma=sigma)
    out = mark_boundaries(img, seg)
    plt.imshow(out)
interact(sp_slic, n_segments=(0,250), compactness=(0,10), sigma=(0,2))
```

```python
def sp_quickshift(kernel_size, max_dist, ratio):
    seg = quickshift(img, kernel_size=kernel_size, max_dist=max_dist, ratio=ratio)
    out = mark_boundaries(img, seg)
    plt.imshow(out)
interact(sp_quickshift, kernel_size=(0,6), max_dist=(0,12), ratio=(0,1.0))
```

```python
def sp_watershed(markers, compactness):
    compactness *= 0.001
    gradient = sobel(rgb2gray(img))
    seg = watershed(gradient, markers=markers, compactness=compactness)
    out = mark_boundaries(img, seg)
    plt.imshow(out)
interact(sp_watershed, markers=(0,500), compactness=(0,2))
```
